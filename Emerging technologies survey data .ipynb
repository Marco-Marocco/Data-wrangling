A good idea is to check the percentage of missing values, and then decide what to do. I will do these check in order:
1. Too many: I will drop the column
2. A good percentage: I will fill the NaN with either the mean if the variable is numeric or the most recurring value if the variable is not numeric
3. Very few: I will drop all the NaN 

```python
df_null=(100* df.isnull().sum()/len(df)).sort_values(ascending=False).to_frame()
df_null=df_null.rename(columns={0:'%missing_data'})
df_null=df_null[df_null['%missing_data']>5]
df_null
```

I do not have too many NaN so I will perform immediately the second task:

```python
for i in df_null.index:
    print(i, ': ',df[i].dtypes)
```

Numeric variables:

```python
list_of_numeric=[]
for i in df_null.index:
    if df[i].dtypes==int or df[i].dtypes==float:
        list_of_numeric.append(i)

list_of_numeric
```

```python
for i in list_of_numeric:
    # filling with the mean of the class
    df[i] = df[i].fillna(df[i].mean())
```

Not numeric variables:

```python
list_of_NOT_numeric=[]
for i in df_null.index:
    if df[i].dtypes==object:
        list_of_NOT_numeric.append(i)

list_of_NOT_numeric
```

```python
for i in list_of_NOT_numeric:
    # filling with the most recurrring of the class
    df[i] = df[i].fillna(df[i].value_counts().index[0])
```

```python
df_null=(100* df.isnull().sum()/len(df)).sort_values(ascending=False).to_frame()
df_null=df_null.rename(columns={0:'%missing_data'})
df_null=df_null[df_null['%missing_data']>0]
df_null
```

I will drop the NaN for the variables with very few NaN: (I can do like that because I do not have other great missing values):

```python
df.dropna(inplace=True)
```

### Normalizing data


There are two columns in the dataset that talk about compensation.

One is "CompFreq". This column shows how often a developer is paid (Yearly, Monthly, Weekly).

The other is "CompTotal". This column talks about how much the developer is paid per Year, Month, or Week depending upon his/her "CompFreq". 

This makes it difficult to compare the total compensation of the developers.

I will create a new column called 'NormalizedAnnualCompensation' which contains the 'Annual Compensation' irrespective of the 'CompFreq', this will make comparison of salaries easy.


```python
df[['CompFreq','CompTotal']].head()
```

```python
df['CompFreq'].unique()
```

```python
df['CompFreq'].value_counts()
```

I will find the multiplier to get the Annual compensation:

```python
def change(frequence):
    if frequence =='Yearly':
        n=1
    if frequence =='Monthly':
        n=12
    if frequence =='Weekly':
        n=52
        
    return n
```

```python
df['multiplier']=df['CompFreq'].apply(change)
```

```python
df.head()
```

```python
df['NormalizedAnnualCompensation']=df['CompTotal']*df['multiplier']
df.head()
```
